# **Scatter plot between age and cholestrol**

```python
sns.scatterplot(data=data,x = 'age', y = 'chol',alpha = 0.5, color = 'green')
plt.xlabel('Age')
plt.ylabel('Cholesterol')
plt.title('Age-Cholesterol Scatter Plot')
```

```python
fig, axes = plt.subplots(1, 2)
sns.scatterplot(
    data=data[data["target"]==1], x="age", y="thalach", color = 'green',alpha = 0.5,
    sizes=(20, 200),ax=axes[0]
)
axes[0].set_title("Have heart disease")
axes[0].set_ylabel("Maximum Heart Rate")
sns.scatterplot(
    data=data[data["target"]==0], x="age", y="thalach", color ='green',alpha = 0.5,
    sizes=(20, 200),ax=axes[1]
)
axes[1].set_title("Does not have Heart disease")
axes[1].set_ylabel("Maximum Heart Rate")
# set the spacing between subplots
plt.subplots_adjust(left=0.0,
                    bottom=0.0, 
                    right=0.9, 
                    top=0.9, 
                    wspace=0.3, 
                    hspace=0.5)
```

```python
fig, axes = plt.subplots(1, 2)
sns.scatterplot(
    data=data[data["target"]==1], x="age", y="chol", color = 'red',alpha = 0.5,
    sizes=(20, 200),ax=axes[0]
)
axes[0].set_title("Have heart disease")
axes[0].set_ylabel("Cholestrol level")
sns.scatterplot(
    data=data[data["target"]==0], x="age", y="chol", color ='red',alpha = 0.5,
    sizes=(20, 200),ax=axes[1]
)
axes[1].set_title("Does not have Heart disease")
axes[1].set_ylabel("Cholestrol level")
# set the spacing between subplots
plt.subplots_adjust(left=0.0,
                    bottom=0.0, 
                    right=0.9, 
                    top=0.9, 
                    wspace=0.3, 
                    hspace=0.5)
```

# **Correlogram**

```python
g = sns.PairGrid(data[["trestbps","chol","thalach","age","oldpeak"]])
g.map(sns.scatterplot)
```

```python
fig,ax = plt.subplots(figsize=(15,15))
sns.heatmap(data.corr(), annot=True, linewidths=.5, fmt= '.1f',ax=ax)
plt.show()
```

# **Tree Map**

```python
data_sub=data.groupby('sex')[['target']].sum()
data_sub
```

```python
sizes=data_sub["target"].tolist()
label=data_sub.index.tolist()
squarify.plot(sizes=sizes, label=label, alpha=0.6 )
plt.axis("off")
plt.show()
```

```python
fig = px.treemap(data, path=["cp","sex"], values="target")
fig.show()
```
